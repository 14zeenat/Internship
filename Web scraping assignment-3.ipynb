#### 2) In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 

```python
# fetching url of each product (excluding sponsored product)
product_url = []
start = 0
end = 3

for page in range(start,end):
    tags = driver.find_elements(By.XPATH,'//div[@class="sg-col-4-of-24 sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-20"]')
    for t in tags:
            url = t.find_elements(By.XPATH,'.//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
            for i in url:
                urls = i.get_attribute('href')
                product_url.append(urls)
        
    time.sleep(5)
    if page<(end-1):
        next = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
        next.click()
    time.sleep(3)
```

```python
# empty lists

product_name = []
brand_name = []
price = []
return_exchange = []
expected_delivery = []
availability = []

# scraping product details
for url in product_url:
    driver.get(url)
    time.sleep(5)
    
    try:
        product_name_tag = driver.find_element(By.ID,"productTitle")
        product_name.append(product_name_tag.text)
    except NoSuchElementException:
        product_name_tag = driver.find_element(By.ID,"title_feature_div")
        product_name.append(product_name_tag.text)
    time.sleep(2)
        
    try:         
        brand= driver.find_element(By.XPATH,'//tr[@class="a-spacing-small po-brand"]/td[2]/span')
        brand_name.append(brand.text) 
    except NoSuchElementException:
        brand_name.append("-")
    time.sleep(2)    
    
    try:
        price_tag= driver.find_element(By.XPATH,'//*[@id="corePriceDisplay_desktop_feature_div"]/div/span[2]')
        price.append(price_tag.text)
    except NoSuchElementException:
        price.append("-")
    time.sleep(2)     
    
    try:
        ret_ex = driver.find_element(By.XPATH,'//*[@id="RETURNS_POLICY"]//div[2]')
        return_exchange.append(ret_ex.text)
    except NoSuchElementException:
        return_exchange.append("-")
    time.sleep(2)
    
    try:
        delivery = driver.find_element(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span[1]')
        expected_delivery.append(delivery.text)
    except NoSuchElementException:
        expected_delivery.append("-")

    try:
        avail = driver.find_element(By.XPATH,'//*[@id="availabilityInsideBuyBox_feature_div"]')
        availability.append(avail.text)
    except NoSuchElementException:
        availability.append("-")
    time.sleep(5)
    
    
```

```python
print(len(product_name),len(expected_delivery),len(return_exchange),len(brand_name),len(availability),len(price),len(product_url))
```

```python
# making dataframe
Guitar = pd.DataFrame({'Product name':product_name,'Brand':brand_name,'Price':price,'Availability':availability,'Expected delivery':expected_delivery,'Return/Exchange':return_exchange,'Product URL':product_url})
Guitar
```

```python
# saving data in csv file
Guitar.to_csv('Guitar.csv')
```

#### 9) Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description. 

```python
# Import requried libraries

import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import time
import warnings
warnings.filterwarnings("ignore")
```

```python
# connecting to the driver
driver = webdriver.Chrome(r"Z:\chromedriver\chromedriver.exe")
```

```python
# opening the page on automated chrome browser
driver.get("https://www.hostelworld.com/")
```

```python
driver.maximize_window()
```

```python
# Enter location in search bar field
location = driver.find_element(By.ID,"search-input-field")
location.send_keys('London')
```

```python
# click on dropdown bar
sel = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]')
sel.click()
```

```python
# clicking search button
search_button = driver.find_element(By.ID,"search-button")
search_button.click()
```

```python
# creating empty lists for storing data
hostel_name = []
dist_from_city = []
private_room = []
dorms = []
hostel_url = []


# scraping and storing required data

    
hostel = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]')           
for h in hostel:
    hostel_name.append(h.text)
    
distance = driver.find_elements(By.XPATH,'//span[@class="description"]//span')
for d in distance:
    dist_from_city.append(d.text.split("-")[1])
        
private = driver.find_elements(By.XPATH,'//div[1][@class="price-col"]')
for i in private:
    try:
        pri = i.find_element(By.XPATH,'div')
        price = pri.text.split(" ")
        if len(price)==2:
            private_room.append(price[1])
        elif len(price)==1:
            private_room.append(price[0])
    except NoSuchElementException:
        private_room.append("-")

dorm_tag = driver.find_elements(By.XPATH,'//div[2][@class="price-col"]')
for j in dorm_tag:
    try:
        dor = j.find_element(By.XPATH,'div')
        dor_price = dor.text.split(" ")
        if len(dor_price)==2:
            dorms.append(dor_price[1])
        elif len(dor_price)==1:
            dorms.append(dor_price[0])
    except NoSuchElementException:
        dorms.append("-")
        
urls = driver.find_elements(By.XPATH,'//h2[@class="title title-6"]//a')
for u in urls:
    hostel_url.append(u.get_attribute('href'))
```

```python
print(len(hostel_name),len(dist_from_city),len(private_room),len(dorms),len(hostel_url))
```

```python
# creating empty lists
rating = []
total_reviews = []
overall_reviews = []
facilities = []
property_description = []

# scraping and storing data from the urls

for url in hostel_url:
    driver.get(url)
    time.sleep(10)
    
    try:
        rating_tag = driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/section/div[6]/div/div[1]/div[1]/div[1]").text
        rating.append(rating_tag.split("\n")[0])
    except NoSuchElementException:
        rating.append("-")
        
    try:
        total_rev = driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/section/div[6]/div/div[1]/div[1]/div[2]/div[2]").text
        total_reviews.append(total_rev.split(" ")[0])
    except NoSuchElementException:
        total_reviews.append("-")
        
    try:
        overall = driver.find_element(By.XPATH,"/html/body/div[3]/div/div/div[2]/section/div[6]/div/div[1]/div[1]/div[2]/div[1]/span").text
        overall_reviews.append(overall)
    except NoSuchElementException:
        overall_reviews.append("-")
        
    try:
        facilities_tag = driver.find_elements(By.XPATH,'//ul[@class="facilities"]')
        fac = []
        for f in facilities_tag:
            sf = f.find_elements(By.XPATH, 'li')
            for k in sf:
                fac.append(k.text)
        facilities.append(tuple(fac))
    except NoSuchElementException:
        facilities.append("-")
        
    try:
        property1 = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[6]/div/div[2]/div/div/div[2]')
        property_description.append(property1.text)
    except NoSuchElementException:
        property_description.append("-")
```

```python
print(len(rating),len(total_reviews),len(overall_reviews),len(facilities),len(property_description))
```

```python
# making dataframe
df_hostel = pd.DataFrame({
    'Hostel name':hostel_name,
    'Distance from city centre':dist_from_city,
    'Ratings':rating,
    'Total reviews':total_reviews,
    'overall reviews':overall_reviews,
    'privates from price':private_room,
    'dorms from price':dorms,
    'facilities':facilities,
    'property description':property_description})
```
