#### Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
##### 1. Brand
##### 2. ProductDescription

#### To scrape the data you have to go through following steps:
##### 1. Go to Flipkart webpage by url :https://www.flipkart.com/
##### 2. Enter “sunglasses” in the search field where “search for products, brands and more” is written and  click the search icon
##### 3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the required data as usual
##### 4. After scraping data from the first page, go to the “Next” Button at the bottom other page , then click on it.
##### 5. Now scrape data from this page asusual
##### 6. Repeat this until you get data for 100sunglasses.


```python
# Import requried libraries
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
# connecting to the driver
driver = webdriver.Chrome(r"Z:\chromedriver\chromedriver.exe")
```

```python
# opening the page on automated chrome brower
driver.get("https://www.flipkart.com/")
```

```python
# cancel the login pop-up
cancel = driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _2doB4z"]')
try:
    cancel.click()
except:
    pass
```

```python
driver.maximize_window()
```

```python
# enter the item name
items = driver.find_element(By.CLASS_NAME,"_3704LK")
items.send_keys('Sunglasses')
```

```python
# click on search
search_4 = driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search_4.click()
```

```python
# creating empty list
brand_name_4 = []
product_details_4 = []
price_4 = []
discount_4 = []
```

```python
start = 0   
end = 3
for page in range(start,end):
    brand_tags_4 = driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')    # scraping brand name
    for i in brand_tags_4:
        brand_name_4.append(i.text)
        brand_4 = brand_name_4[0:100]
      
    product_tags_4 = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')      # scraping product description
    for j in product_tags_4:
        product_details_4.append(j.text)
        product_4 = product_details_4[0:100]
    
    price_tags_4 = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')     # scraping price
    for k in price_tags_4:
        price_4.append(k.text)
        pri_4 = price_4[0:100]         
       
    discount_tags_4 = driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')   # scraping discount
    for l in discount_tags_4:
        discount_4.append(l.text)
        dis_4 = discount_4[0:100]
    time.sleep(2)
        
    if page < (end-1):    
        next_button = driver.find_element(By.XPATH, "//span[text()='Next']")    # click on next button
        next_button.click()
    time.sleep(2)
```

```python
print(len(brand_4),len(product_4),len(pri_4),len(dis_4))
```

```python
# making dataframe
df_sunglasses = pd.DataFrame({'Brand':brand_4,'product_discription':product_4,'Price':pri_4,'Discount':dis_4})
df_sunglasses
```

#### Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
##### The above task will be done in following steps:
##### 1. First get the webpagehttps://www.azquotes.com/
##### 2. Click on TopQuote
##### 3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
# Import requried libraries
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
# connecting to the driver
driver = webdriver.Chrome(r"Z:\chromedriver\chromedriver.exe")
```

```python
# opening the page on automated chrome brower
driver.get("https://www.azquotes.com/")
```

```python
driver.maximize_window()
```

```python
#  selecting top quote
top_quote = driver.find_element(By.XPATH,'//a[@href="/top_quotes.html"]')
top_quote.click()
```

```python
# empty lists
Quote = []
Author = []
Types = []
```

```python
start = 0
end = 10

for page in range(start,end):
    quote_8 = driver.find_elements(By.XPATH, '//a[@class="title"]')         # scraping quote
    for i in quote_8:
        Quote.append(i.text)
        
    author_8 = driver.find_elements(By.XPATH, '//div[@class="author"]')     # scraping author name
    for j in author_8:
        Author.append(j.text)
        
    ToQ = driver.find_elements(By.XPATH,'//div[@class="tags"]')            # scraping type of quotes
    for k in ToQ:
        Types.append(k.text)
    time.sleep(3)
        
    if page < (end-1):    
        Next = driver.find_element(By.XPATH,'//li[@class="next"]')         # click on the next button
        Next.click()
    time.sleep(3)
```

```python
print(len(Quote),len(Author),len(Types))
```

```python
# making dataframe
df_Quotes = pd.DataFrame({'Quote':Quote,'Author':Author,'Type of Quotes':Types})
```

```python
# empty list
cars = []
car_price = []
    
# scraping car price
car_name = driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in car_name[0:50]:
    cars.append(i.text)
    
# scraping car price
price = driver.find_elements(By.TAG_NAME,"strong")
for j in price:
    n = j.text.split(':')
    if len(n)==2:
        car_price.append(n[1])
    elif n[0]=='':
        n.pop()
    elif len(n)==1:
        car_price.append(n[0])   
```

```python
print(len(cars),len(car_price))
```

```python
# making DataFrame
df_cars = pd.DataFrame({'Car_name':cars,'Price':car_price})
df_cars
```
